# Data Science Model Set-Up

### Import Libraries

```
import pandas as pd
import os
import xlrd
import openpyxl
```

```
df_gv = pd.read_excel('12_DICIEMBRE_2020.xlsx', sheet_name ='GASTOS VARIOS', skiprows=5)
df_gv['SHEET'] = 'GASTOS VARIOS'
```

```
df_contratist = pd.read_excel('12_DICIEMBRE_2020.xlsx', sheet_name ='CONTRATISTAS Y FDO FED', skiprows=5)
df_contratist['SHEET'] = 'CONTRATISTAS Y FDO FED'
```

```
df_service = pd.read_excel('12_DICIEMBRE_2020.xlsx', sheet_name ='SERV PPROF', skiprows=5)
df_service['SHEET'] = 'SERV PPROF'
```

```
df_communication = pd.read_excel('12_DICIEMBRE_2020.xlsx', sheet_name ='COMUNICACION', skiprows=5)
df_communication['SHEET'] = 'COMUNICACION'
```

```
df_personal = pd.read_excel('12_DICIEMBRE_2020.xlsx', sheet_name ='SERV. PERS.', skiprows=5)
df_personal['SHEET'] = 'SERV. PERS.'
```

```
df_buy_dec = pd.concat([df_gv, df_contratist, df_service, df_communication, df_personal])
```

### Let's now check the concatenate information:

```
df_buy_dec.head(5)
```

### Let's now deal with Null-Values if there is any

```
df_buy_dec_com = df_buy_dec.dropna(how='any')
```
